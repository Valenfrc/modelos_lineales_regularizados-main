```python
#analisis multivariante
#numerico-numerico

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Percent of adults with a high school diploma only 2014-18': [50, 60, 70, 80, 90, 75, 85, 65, 55, 45],
    'diabetes_prevalence': [10, 12, 14, 16, 18, 15, 17, 13, 11, 9],
    'duration': [2, 3, 4, 5, 6, 4, 5, 3, 2, 1],
    'age': [60, 65, 70, 75, 80, 72, 78, 68, 62, 58],
    'Percent of Population Aged 60+': [20, 25, 30, 35, 40, 32, 38, 28, 22, 18],
    'Heart disease_Upper 95% CI': [10, 12, 14, 16, 18, 15, 17, 13, 11, 9]
}
df3 = pd.DataFrame(data)

fig, axes = plt.subplots(3, 2, figsize=(10, 15))  # Adjusted to 3x2

sns.regplot(ax=axes[0, 0], data=df3, x='Percent of adults with a high school diploma only 2014-18', y='diabetes_prevalence')
sns.heatmap(df3[["duration", "age"]].corr(), annot=True, fmt=".2f", ax=axes[1, 0], cbar=False)

sns.regplot(ax=axes[0, 1], data=df3, x='diabetes_prevalence', y='Percent of Population Aged 60+')
sns.heatmap(df3[['diabetes_prevalence', 'Percent of Population Aged 60+']].corr(), annot=True, fmt=".2f", ax=axes[1, 1], cbar=False)

sns.regplot(ax=axes[2, 0], data=df3, x='Heart disease_Upper 95% CI', y='diabetes_prevalence')
sns.heatmap(df3[['diabetes_prevalence', 'Heart disease_Upper 95% CI']].corr(), annot=True, fmt=".2f", ax=axes[2, 1], cbar=False) #added heatmap to position [2,1]

plt.tight_layout()
plt.show()

```

```python
for column in df.columns:
    if df[column].dtype=='object':
        df[column +'_num']=pd.factorize(df[column])[0]
```

```python
df2= df.drop(['COUNTY_NAME','STATE_NAME'], axis= 1 )
```

```python
for column in df2.columns:
    if 'diabetes' in column:
        if column!='diabetes_prevalence':
            print(column)
            df2.drop(column,axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x=df2.drop(['diabetes_prevalence'],axis=1)
y=df2['diabetes_prevalence']

X_train, X_test, Y_train, Y_test= train_test_split(x,y,test_size=0.2,random_state=42)

X_train
```

```python
from sklearn.preprocessing import  StandardScaler

scaler=StandardScaler()
scaler.fit(X_train,Y_train)
xtrain_escalados= scaler.transform(X_train)
xtest_escalados=scaler.transform(X_test)

X_train_esc=pd.DataFrame(xtrain_escalados, index=X_train.index,columns=X_train.columns)
X_test_esc=pd.DataFrame(xtest_escalados, index=X_test.index,columns=X_test.columns)

```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

model_rl = LinearRegression()
model_lasso = Lasso()
model_ridge = Ridge()

model_rl.fit(X_train_esc, Y_train)
model_lasso.fit(X_train_esc, Y_train)
model_ridge.fit(X_train_esc, Y_train)
```

```python
y_rl_pred = model_rl.predict(X_test_esc)
y_lasso_pred = model_lasso.predict(X_test_esc)
y_ridge_pred = model_ridge.predict(X_test_esc)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print(f"El error de la Regresion Lineal es{mean_squared_error(Y_test,y_rl_pred)} y el R2 es de {r2_score(Y_test,y_rl_pred)}")
print(f"El error de la Regresion Lasso es{mean_squared_error(Y_test,y_lasso_pred)} y el R2 es de {r2_score(Y_test,y_lasso_pred)}")
print(f"El error de la Regresion Ridge es{mean_squared_error(Y_test,y_ridge_pred)} y el R2 es de {r2_score(Y_test,y_ridge_pred)}")
```
