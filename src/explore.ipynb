```python
#empezare a analizar las variables categoricas

import matplotlib.pyplot as plt
import seaborn as sns

c=df.groupby('COUNTY_NAME').size().reset_index()
s=df.groupby('STATE_NAME').size().reset_index()
print('c','s')
```

```python
C2=c.sort_values(by=0,ascending=False)[:10]
```

```python

import matplotlib.pyplot as plt
import pandas as pd 

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 10)) 



axes[0].bar(C2.COUNTY_NAME, C2[0])
axes[0].set_title("COUNTY_NAME")

axes[1].bar(s.STATE_NAME, s.iloc[:, 0])
axes[1].set_title("STATE_NAME")

# Reduce font size and adjust spacing
axes[0].tick_params(axis='x', rotation=90, labelsize=8)  # Smaller font
axes[1].tick_params(axis='x', rotation=90, labelsize=8)  # Smaller font

plt.tight_layout()
plt.show()
```

```python

df=df.rename(columns={'0-9 y/o % of total pop': '0-9 total pop'})
```

```python
#Ahora analizaremos las variables numericas
import seaborn as sns
fig, axes = plt.subplots(12, 2, figsize=(15, 30))

variables_numericas = ['fips','TOT_POP','0-9','0-9 total pop','19-Oct','CKD_prevalence','CKD_Lower 95% CI','CKD_Upper 95% CI','CKD_number','Urban_rural_code']

for i, variable in enumerate(variables_numericas):
    # Histograma
    sns.histplot(ax=axes[i, 0], data=df, x=variable, kde=True)  # kde para suavizar la curva
    axes[i, 0].set_title(f"Histograma de {variable}")
    axes[i, 0].set_xlabel(variable)
    axes[i, 0].set_ylabel("Frecuencia")

    # Boxplot
    sns.boxplot(ax=axes[i, 1], data=df, x=variable)
    axes[i, 1].set_title(f"Boxplot de {variable}")
    axes[i, 1].set_xlabel(variable)
    axes[i, 1].set_ylabel("Valor")  # Ajusta la etiqueta del eje y

plt.tight_layout()
plt.show()

#el histplot lo estoy usando para enterarme de los datos recurrentes
#el boxplot lo estoy utilizando para los valores atipicos
```
