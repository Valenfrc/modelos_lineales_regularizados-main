```python

df=df.rename(columns={'0-9 y/o % of total pop': '0-9 total pop'})
```

```python
#analisis numerico

import seaborn as sns
import matplotlib.pyplot as plt


variables_numericas = [
    'Active Physicians per 100000 Population 2018 (AAMC)',
    'Total Hospitals (2019)',
    'ICU Beds_x',
    'anycondition_prevalence',
    'Obesity_prevalence',
    'Heart disease_prevalence',
    'diabetes_prevalence',
    'COPD_prevalence',
    'Total nurse practitioners (2019)',
    'Total physician assistants (2019)',
    'TOT_POP',
    'Percent of Population Aged 60+',
    'MEDHHINC_2018',
    'Unemployment_rate_2018',
    'PCTPOVALL_2018',
    '% White-alone',
    '% Black-alone',
    '% Asian-alone',
    "Percent of adults with a bachelor's degree or higher 2014-18", 
    'Urban_rural_code'
]
fig, axes = plt.subplots(20, 2, figsize=(15, 60))

for i, variable in enumerate(variables_numericas):
    # Histograma
    sns.histplot(ax=axes[i, 0], data=df, x=variable, kde=True)
    axes[i, 0].set_title(f"Histograma de {variable}")
    axes[i, 0].set_xlabel(variable)
    axes[i, 0].set_ylabel("Frecuencia")

    # Boxplot
    sns.boxplot(ax=axes[i, 1], data=df, x=variable)
    axes[i, 1].set_title(f"Boxplot de {variable}")
    axes[i, 1].set_xlabel(variable)
    axes[i, 1].set_ylabel("Valor")

plt.tight_layout()
plt.show()

#Los recursos sanitarios seleccionados ofrecen una visión completa de la disponibilidad y el estado de los servicios médicos, así como de la prevalencia de enfermedades clave
```

```python
#analisis multivariante
#numerico-numerico

#anycondition_prevalence. Esta variable representa la prevalencia de "cualquier condición de salud crónica". Esto la convierte en un indicador amplio y general del estado de salud de una población.
fig,axis= plt.subplots(4,2, figsize=(10,20))

sns.regplot(ax = axis[0, 0], data = df, x = "anycondition_prevalence", y = "age")
sns.heatmap(df[["duration", "age"]].corr(), annot = True, fmt = ".2f", ax = axis[1, 0], cbar = False)

sns.regplot(ax=axis[0,1],data=df,x="duration",y="campaign")
sns.heatmap(df[["duration", "campaign"]].corr(), annot = True, fmt = ".2f", ax = axis[1,1 ], cbar = False)

sns.regplot(ax = axis[2, 0], data = df, x = "previous", y = "duration")
sns.heatmap(df[["duration", "previous"]].corr(), annot = True, fmt = ".2f", ax = axis[3, 0], cbar = False)

plt.show

```
